```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.ticker import FormatStrFormatter
data = pd.read_csv('data_nadoll.csv')
```

```python
palette = ['#FFFF00','#A8CF45','#0080FF','#FF7EC0', '#F58634']
sns.set_palette(palette, n_colors=5, color_codes=True)
sns.color_palette()
```

```python
plt.figure(figsize=(6,7))
plt.grid(True, linestyle=':', alpha=0.5) 
nadoll=sns.scatterplot(x="tiv", y="almn", s=100, data=data, hue="Localidad", palette=['#A8CF45', '#F58634', '#FFFF00', '#FF7EC0'], 
                       alpha=0.8, edgecolor="white")
#s: escalar
plt.xscale('log')
plt.yscale('log')
plt.gca().xaxis.set_major_formatter(FormatStrFormatter('%.2f'))
plt.gca().yaxis.set_major_formatter(FormatStrFormatter('%.1f'))
plt.legend(title='Localidad', loc='lower right', ncol=2)
plt.xlim(0.0001,10)
plt.ylim(0.001,10)
plt.xlabel("Ti+V(wt.%)")
plt.ylabel("Al+Mn(wt.%)")
#plt.savefig("abc.png", dpi=300)
plt.show()
```
