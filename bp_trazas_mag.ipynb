```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.ticker import FormatStrFormatter
trazas = pd.read_csv('traza.csv')
```

```python
print (trazas.head(6))
```

```python
palette = ['#FFFF00','#A8CF45','#0080FF','#FF7EC0', '#F58634']
sns.set_palette(palette, n_colors=5, color_codes=True)
sns.color_palette()
```

```python
bp=sns.boxplot(data=trazas) #, x="Sample", y="Mg25_ppm_mean", width=0.2) #hue='sample') #color=white 
bp.set_xticklabels(bp.get_xticklabels(), rotation=90)
#sns.color_palette()
plt.yscale('log')
plt.show()
```

```python
mg=sns.boxplot(data=trazas,x="Sample", y="Cu", hue='Locality', palette=['#A8CF45','#F58634','#FFFF00','#FF7EC0'], saturation=1, width=0.8) #hue='sample') #color=white 
mg.set_xticklabels(mg.get_xticklabels(), rotation=90)
#sns.color_palette()
plt.yscale('log')
mg.yaxis.set_major_formatter(FormatStrFormatter('%.2f'))
plt.title("Cu")
plt.xlabel("Muestra")
plt.ylabel("ppm")
plt.legend(title='Localidad', loc='lower right', ncol=2)
plt.show()
```

```python
al=sns.boxplot(data=trazas,x="Sample", y="Al", hue='Locality', palette=['#A8CF45','#F58634','#FFFF00','#FF7EC0'], saturation= 1,
               width=0.3, dodge=False,fliersize=3,linewidth=1)
               #dodge=0.01) #hue='sample') #color=white 
al.set_xticklabels(al.get_xticklabels(), rotation=90)
plt.yscale('log')
plt.title("Al")
plt.xlabel("Muestra")
plt.ylabel("ppm")
plt.legend(title='Localidad', loc='lower right', ncol=2)
plt.show()
```
